# Implementation of K-Mean Clustering

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler,StandardScaler

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Iris.csv')
df
```

```python
x = df.drop(['Id','Species'], axis=1)
x
```

# Preprocessing Scaling

```python
std_scaler = StandardScaler()
x_scaled = std_scaler.fit_transform(x)
x_df = pd.DataFrame(x_scaled, columns=x.columns)
x_df
```

# model Training

```python
Kmean_model = KMeans(n_clusters=3)
Kmean_model.fit(x_df)
pred_var = Kmean_model.fit_predict(x_df)
pred_var
```

```python
class1=np.where(pred_var==0)
```

```python
x.iloc[class1]
```

```python
wcss=Kmean_model.inertia_
wcss
```

```python
wcsslist=[]
for k in range(1,11):
    Kmean_model=KMeans(n_clusters=k)
    Kmean_model.fit(x_df)
    wcss=Kmean_model.inertia_
    wcsslist.append(wcss)
    print(f'k=={k} and wcss == {wcss}')
```

```python
k_val=range(1,11)
plt.plot(k_val,wcsslist)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
```
