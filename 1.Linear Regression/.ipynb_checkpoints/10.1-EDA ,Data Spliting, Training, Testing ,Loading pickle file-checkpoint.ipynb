# data spliting

```python
x=df.drop('price',axis=1)
y=df['price']
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

## model Training

```python
df_train=LinearRegression()
df_train.fit(x_train,y_train)
```

## model_test

```python
y_pred=df_train.predict(x_test)
```

```python
y_pred
```

```python
y_pred_train=df_train.predict(x_train)
```

```python
y_pred_train
```

## evaluation on training data

```python
y_pred_train=df_train.predict(x_train)

mse=mean_squared_error(y_train,y_pred_train)
print('mean squared error :',mse)

rmse=np.sqrt(mse)
print('root mean squared error :',rmse)

mae=mean_absolute_error(y_train,y_pred_train)
print('mean absolute squared error :',mae)

R2=r2_score(y_train,y_pred_train)
print('coeficient of determination r2 :',R2)

adj_R2=1-(1-R2)*(x_train.shape[0]-1)/(x_train.shape[0]-13-1)
print('adjusted mean squared :',adj_R2)

```

## evaluation on testing data

```python
y_pred=df_train.predict(x_test)

mse=mean_squared_error(y_test,y_pred)
print('mean squared error :',mse)

rmse=np.sqrt(mse)
print('root mean squared error :',rmse)

mae=mean_absolute_error(y_test,y_pred)
print('mean absolute squared error :',mae)

R2=r2_score(y_test,y_pred)
print('coeficient of determination r2 :',R2)

adj_R2=1-(1-R2)*(x_test.shape[0]-1)/(x_test.shape[0]-13-1)
print('adjusted mean squared :',adj_R2)
```

## normality of residual

```python
Residual=y_train-y_pred_train
```

```python
Residual
```

```python
sns.kdeplot(Residual,fill=True)
```

```python
sns.displot(Residual)
```

```python
mean=df['price'].mean()
print('mean is ',mean)

median=df['price'].median()
print('median is ',median)

mode=df['price'].mode()
print('mode is ',mode)

skew=df['price'].skew()
print('skewness is :',skew)
```

```python
from scipy.stats import skew
```

```python
skewness=skew(df['price'])
skewness
```

```python
import statsmodels.api as sm
```

```python
sm.qqplot(Residual,line='45',fit=True)
```

```python
from scipy.stats import shapiro
```

```python
_,p_val=shapiro(Residual)
print('p value:',p_val)

if p_val>=0.05:
    print('null hypothesis is true')
    print('data is normally distributed')
    
else:
    print('Alternate Hypothesis is true')
    print('Data is not Normally distributed')
```

```python
from scipy.stats import kstest
```

```python
_,p_val=kstest(Residual,'norm')
print('p value:',p_val)

if p_val>=0.05:
    print('null hypothesis is true')
    print('data is normally distributed')
    
else:
    print('Alternate Hypothesis is true')
    print('Data is not Normally distributed')
```

```python
from scipy.stats import normaltest
```

```python
_,p_val=normaltest(Residual)
print('p value:',p_val)

if p_val>=0.05:
    print('null hypothesis is true')
    print('data is normally distributed')
    
else:
    print('Alternate Hypothesis is true')
    print('Data is not Normally distributed')
```
