```python
import numpy as np
import pandas as pd

from sklearn.impute import KNNImputer
```

```python
df=pd.DataFrame({'A':[1,4,5,7,6,np.nan,5,3,7,8],
                 'B':[10,np.nan,12,18,14,13,11,10,13,11],
                 'C':[30,40,45,34,52,50,40,58,41,32]})
df
```

```python
knn_imputer=KNNImputer(n_neighbors=3)
array=knn_imputer.fit_transform(df)
df1=pd.DataFrame(array,columns=df.columns)
df1
```
